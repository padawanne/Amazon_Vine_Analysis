```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read vine_table 
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
vine_df.info()
```

```python
# Display Dataframe
vine_df
```

```python
# Filter data total_votes >= 20
total_votes_count = vine_df[vine_df['total_votes'] >= 20]
total_votes_count.head()
```

```python
# Find helpful_votes
new_vine_df = total_votes_count.loc[total_votes_count.helpful_votes / total_votes_count.total_votes >= 0.5]
new_vine_df.head()
```

```python
# Reviews Vine Program
paid_reviews = new_vine_df[new_vine_df['vine'] == 'Y']
paid_reviews.head()
```

```python
# Reviews outside of Vine Program
unpaid_reviews = new_vine_df[new_vine_df['vine'] == 'N']
unpaid_reviews.head()
```

```python
# Count total reviews Vine
total_reviews_Vine = len(paid_reviews)
total_reviews_Vine
```

```python
# Count total reviews not Vine / unpaid
total_reviews_unpaid = len(unpaid_reviews)
total_reviews_unpaid
```

```python
# 5 Star reviews Vine 
vine_five_star = len(paid_reviews[paid_reviews['star_rating'] == 5])
vine_five_star
```

```python
# 5 Star reviews not Vine / unpaid
unpaid_five_star = len(unpaid_reviews[unpaid_reviews['star_rating'] == 5])
unpaid_five_star
```

```python
# Percentage 5 Star reviews Vine
vine_five_star/total_reviews_Vine
```

```python
# Percentage 5 Star reviews not Vine / unpaid
unpaid_five_star/total_reviews_unpaid
```
